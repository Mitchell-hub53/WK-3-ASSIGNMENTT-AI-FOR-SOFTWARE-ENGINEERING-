```python
# 2C_spacy_ner_sentiment.ipynb
# Install spaCy and the small English model (run once)
!pip install -q spacy
!python -m spacy download en_core_web_sm

import spacy
from spacy.pipeline import EntityRuler
import re

# Load model
nlp = spacy.load("en_core_web_sm")

# Add custom EntityRuler to capture product/brand patterns (optional)
ruler = nlp.add_pipe("entity_ruler", before="ner")
patterns = [
    {"label": "PRODUCT", "pattern": [{"LOWER": "echo"}, {"LOWER": "dot"}]},   # example: "Echo Dot"
    {"label": "PRODUCT", "pattern": "iPhone 12"},
    {"label": "BRAND", "pattern": "Apple"},
    {"label": "BRAND", "pattern": "Samsung"},
    {"label": "BRAND", "pattern": "Sony"},
    # Add more patterns relevant to your dataset
]
ruler.add_patterns(patterns)

# Example reviews (replace with sample Amazon reviews or a small CSV)
reviews = [
    "I bought the Echo Dot and it works great. Amazon's Alexa responds quickly.",
    "The iPhone 12 camera is amazing. Apple did a great job.",
    "Terrible battery life on the Samsung TV I purchased last week. Very disappointed.",
    "Great sound quality from the Sony headphones, highly recommend!"
]

# Simple sentiment lexicon (rule-based)
positive_words = {"great","amazing","good","excellent","recommend","love","works","quickly","best","satisfied","awesome"}
negative_words = {"terrible","disappointed","bad","worst","poor","awful","hate","slow","slowly","problem"}

def rule_sentiment(text):
    # naive sentiment: count positive vs negative stems
    tokens = re.findall(r"\w+", text.lower())
    pos = sum(1 for t in tokens if t in positive_words)
    neg = sum(1 for t in tokens if t in negative_words)
    if pos > neg:
        return "positive"
    elif neg > pos:
        return "negative"
    else:
        return "neutral"

# Process and extract entities + sentiment
for review in reviews:
    doc = nlp(review)
    ents = [(ent.text, ent.label_) for ent in doc.ents]
    sentiment = rule_sentiment(review)
    print("REVIEW:", review)
    print("ENTITIES:", ents)
    print("SENTIMENT:", sentiment)
    print("---")
```
